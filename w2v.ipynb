# creative_id

```python
import pandas as pd
import gensim
```

```python
train_seq = pd.read_csv('/DATA/zhouyongyu/taac/user_creative_seq.csv')
```

```python
test_seq = pd.read_csv('/DATA/zhouyongyu/taac/test_creative_seq.csv')
```

```python
train_seq.head()
```

```python
train_creative_id = train_seq['creative_id'].tolist()
```

```python
test_creative_id = test_seq['creative_id'].tolist()
```

```python
all_sentences = []
from ast import literal_eval
from tqdm import tqdm
```

```python
for seq in tqdm(train_creative_id):
    all_sentences.append([str(item) for item in literal_eval(seq)])
```

```python
for seq in tqdm(test_creative_id):
    all_sentences.append([str(item) for item in literal_eval(seq)])
```

```python
print(all_sentences[:10])
```

```python
model = gensim.models.Word2Vec(all_sentences, size=32)
model.save('/DATA/zhouyongyu/model_save/w2v_model/creative_32')
```

```python
model = gensim.models.Word2Vec(all_sentences, size=64)
model.save('/DATA/zhouyongyu/model_save/w2v_model/creative_64')
```

```python
model = gensim.models.Word2Vec(all_sentences, min_count=3, size=128)
model.save('/DATA/zhouyongyu/model_save/w2v_model/creative_128')
```

```python
model = gensim.models.Word2Vec(all_sentences, min_count=1, size=256, sg=1, iter=5, window=10)
model.save('/DATA/zhouyongyu/model_save/w2v_model/creative_256')
```

```python
# 引入日志配置
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
model = gensim.models.Word2Vec(all_sentences, min_count=1, size=300, sg=1, iter=5, window=20, negative=10)
model.save('/DATA/zhouyongyu/model_save/w2v_model/creative_300')
```

```python
model = gensim.models.Word2Vec(all_sentences, min_count=1, size=300, sg=1, iter=5, window=150, negative=10, workers=4)
model.save('/DATA/zhouyongyu/model_save/w2v_model/creative_300_w150')
```
